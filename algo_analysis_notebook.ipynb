```python
# =======================
# CELL 1 — IMPORTS
# =======================

import time
import numpy as np
import matplotlib.pyplot as plt

print("Libraries Loaded Successfully!")
```

**Fibonacci Functions (Recursive + DP)**

```python
# =======================
# CELL 2 — FIBONACCI (RECURSIVE + DP)
# =======================

# Recursive Fibonacci
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

# Dynamic Programming Fibonacci
def fib_dp(n):
    if n <= 1:
        return n
    dp = [0]*(n+1)
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

print("Fibonacci Functions Ready!")
```

**Fibonacci Time Comparison + Combined Graph**

```python
# =======================
# CELL 3 — FIBONACCI TIME ANALYSIS GRAPH
# =======================

n_values = [5, 10, 15, 20, 25, 30]

recursive_time = []
dp_time = []

for n in n_values:
    # Recursive Time
    start = time.perf_counter()
    fib_recursive(n)
    recursive_time.append(time.perf_counter() - start)

    # DP Time
    start = time.perf_counter()
    fib_dp(n)
    dp_time.append(time.perf_counter() - start)

# Combined Graph
plt.figure(figsize=(8,5))
plt.plot(n_values, recursive_time, marker='o', label="Recursive Fibonacci")
plt.plot(n_values, dp_time, marker='o', label="DP Fibonacci")

plt.xlabel("n")
plt.ylabel("Execution Time (seconds)")
plt.title("Fibonacci: Recursive vs DP (Combined Graph)")
plt.grid(True)
plt.legend()
plt.show()

print("Fibonacci Graph Generated!")
```

**Sorting Algorithms**

```python
# =======================
# CELL 4 — SORTING ALGORITHMS
# =======================

# Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i=j=k=0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k]=L[i]; i+=1
            else:
                arr[k]=R[j]; j+=1
            k+=1
        while i < len(L):
            arr[k]=L[i]; i+=1; k+=1
        while j < len(R):
            arr[k]=R[j]; j+=1; k+=1
    return arr

# Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left  = [x for x in arr if x < pivot]
    mid   = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + mid + quick_sort(right)

# Bubble Sort
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1], arr[j]
    return arr

# Selection Sort
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_i = i
        for j in range(i+1,n):
            if arr[j] < arr[min_i]:
                min_i = j
        arr[i],arr[min_i] = arr[min_i],arr[i]
    return arr

# Insertion Sort
def insertion_sort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and key < arr[j]:
            arr[j+1] = arr[j]
            j-=1
        arr[j+1] = key
    return arr

print("Sorting Algorithms Ready!")
```

**Sorting Time Analysis + Combined Graph**

```python
# =======================
# CELL 5 — SORTING TIME COMPARISON GRAPH
# =======================

input_sizes = [100, 300, 500, 700, 900]

merge_t = []
quick_t = []
bubble_t = []
select_t = []
insert_t = []

def measure_time(func, arr):
    start = time.perf_counter()
    func(arr)
    return time.perf_counter() - start

for n in input_sizes:
    arr = np.random.randint(1,1000,n).tolist()

    merge_t.append(measure_time(merge_sort, arr.copy()))
    quick_t.append(measure_time(quick_sort, arr.copy()))
    bubble_t.append(measure_time(bubble_sort, arr.copy()))
    select_t.append(measure_time(selection_sort, arr.copy()))
    insert_t.append(measure_time(insertion_sort, arr.copy()))

# Combined Graph
plt.figure(figsize=(10,6))

plt.plot(input_sizes, merge_t, marker='o', label="Merge Sort")
plt.plot(input_sizes, quick_t, marker='o', label="Quick Sort")
plt.plot(input_sizes, bubble_t, marker='o', label="Bubble Sort")
plt.plot(input_sizes, select_t, marker='o', label="Selection Sort")
plt.plot(input_sizes, insert_t, marker='o', label="Insertion Sort")

plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.title("Sorting Algorithms — Combined Performance Graph")
plt.grid(True)
plt.legend()
plt.show()

print("Sorting Performance Graph Generated!")
```

**Binary Search + Time Graph**

```python
# =======================
# CELL 4 — BINARY SEARCH + GRAPH
# =======================

# Binary Search Function
def binary_search(arr, target):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1


# Function to measure time
import time
def measure_time_bs(arr, target):
    start = time.perf_counter()
    binary_search(arr, target)
    return time.perf_counter() - start


# Test with growing input sizes
sizes = [1000, 5000, 10000, 20000, 30000, 50000]
times = []

for n in sizes:
    arr = list(range(n))       # sorted array for binary search
    target = n - 1             # last element
    t = measure_time_bs(arr, target)
    times.append(t)


# Plot Graph
plt.figure(figsize=(8,5))
plt.plot(sizes, times, marker='o', linestyle='-', color='blue')

plt.title("Binary Search Performance")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.grid(True)
plt.show()

print("Binary Search Graph Generated Successfully!")
```

 **BINARY SEARCH — TIME PROFILING + MEMORY PROFILING + GRAPH**

```python
# ============================================================
# BINARY SEARCH — TIME PROFILING + MEMORY PROFILING + GRAPH
# ============================================================

import time
import numpy as np
import matplotlib.pyplot as plt

# Install memory_profiler if not already installed
!pip install memory_profiler

from memory_profiler import memory_usage


# ------------------------------------------------------------
# Binary Search Function (Iterative)
# ------------------------------------------------------------
def binary_search(arr, target):
    """
    Perform Binary Search on a sorted array.

    Parameters:
        arr (list): Sorted list of numbers.
        target (int): Value to search.

    Returns:
        int: Index of target, -1 if not found.
    """
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid

        elif arr[mid] < target:
            low = mid + 1

        else:
            high = mid - 1

    return -1


# ------------------------------------------------------------
# FUNCTION: Measure Time Taken by Binary Search
# ------------------------------------------------------------
def time_profile(arr, target):
    """
    Returns time taken by binary search in seconds.
    """
    start = time.perf_counter()
    binary_search(arr, target)
    end = time.perf_counter()
    return end - start


# ------------------------------------------------------------
# FUNCTION: Measure Memory Used by Binary Search
# ------------------------------------------------------------
def memory_profile(arr, target):
    """
    Uses memory_profiler to measure memory usage during search.
    Returns memory difference in MB.
    """
    mem_before = memory_usage()[0]
    binary_search(arr, target)
    mem_after = memory_usage()[0]
    return mem_after - mem_before


# ------------------------------------------------------------
# EXPERIMENT: Time & Memory for Growing Input Sizes
# ------------------------------------------------------------
sizes = [1_000, 5_000, 10_000, 20_000, 40_000, 80_000]

time_results = []
memory_results = []

for n in sizes:
    arr = list(range(n))      # sorted list for binary search
    target = n - 1            # best case: last element

    # Time profiling
    t = time_profile(arr, target)
    time_results.append(t)

    # Memory profiling
    m = memory_profile(arr, target)
    memory_results.append(m)


# ------------------------------------------------------------
# GRAPH 1: Execution Time vs Input Size
# ------------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(sizes, time_results, marker='o', linewidth=2)
plt.title("Binary Search — Time Profiling")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.show()


# ------------------------------------------------------------
# GRAPH 2: Memory Usage vs Input Size
# ------------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(sizes, memory_results, marker='o', color='red', linewidth=2)
plt.title("Binary Search — Memory Profiling")
plt.xlabel("Input Size (n)")
plt.ylabel("Memory Used (MB)")
plt.grid(True)
plt.show()


print("✔ Binary Search Time & Memory Profiling Completed!")
```
