```python
#Fibonacci Recursive
import time
import matplotlib.pyplot as plt
import sys

# Recursive Fibonacci Function
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# Values of n to test (changed)
n_values = list(range(1, 36, 2))  # Test odd numbers from 1 to 35
execution_times = []
space_usage = []

for n in n_values:
    # Measuring execution time
    start = time.time()
    fibonacci_recursive(n)
    end = time.time()
    execution_times.append(end - start)

    # Estimating space usage
    frame_size = sys.getsizeof(n)  # size of each recursive frame
    space_usage.append(n * frame_size)

# Plotting results
plt.figure(figsize=(8, 5))
plt.plot(n_values, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(n_values, space_usage, marker='o', color='green', label='Space (bytes)')
plt.title('Time vs Space Complexity of Recursive Fibonacci')
plt.xlabel('n (input size)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Print analysis
print("\n--- Recursive Fibonacci Analysis ---")
print("Input/Output:")
print("  Input: n (integer, position in Fibonacci sequence)")
print("  Output: nth Fibonacci number")
print("\nTime Complexity:")
print("  Best Case: O(1) when n = 0 or 1")
print("  Worst Case: O(2^n) due to repeated recursive calls")
print("  Average Case: O(2^n) (same as worst case)")
print("\nSpace Complexity:")
print("  O(n) because of recursion stack depth")
```

```python
#Fibonacci Dynamic
import time
import matplotlib.pyplot as plt
import sys

# Dynamic Programming Fibonacci
def fibonacci_dp(n):
    if n <= 1:
        return n
    dp = [0, 1]
    for i in range(2, n+1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

# Changed values
n_values = [100, 1000, 5000, 10000, 20000, 50000]
execution_times = []
space_usage = []

for n in n_values:
    start = time.time()
    fibonacci_dp(n)
    end = time.time()

    execution_times.append(end - start)

    frame_size = sys.getsizeof(n)
    space_usage.append(n * frame_size)

# Plotting results
plt.figure(figsize=(8, 5))
plt.plot(n_values, execution_times, marker='o', color='red', label='Time (seconds)')
plt.plot(n_values, space_usage, marker='o', color='purple', label='Space (bytes)')
plt.title('Time vs Space Complexity of DP Fibonacci')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Merge Sort
import time
import matplotlib.pyplot as plt
import sys
import random

# ----------------------------
# Merge Sort Implementation
# ----------------------------
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# ----------------------------
# Profiling Merge Sort
# ----------------------------
input_sizes = [100, 500, 1000, 2000, 5000, 10000]  # Changed values
execution_times = []
space_usage = []

for size in input_sizes:
    arr = [random.randint(1, 10000) for _ in range(size)]

    # Measure execution time
    start = time.time()
    merge_sort(arr)
    end = time.time()
    execution_times.append(end - start)

    # Space usage (approx: n * size of element)
    space_usage.append(size * sys.getsizeof(arr[0]))

# ----------------------------
# Plotting Graph
# ----------------------------
plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='green', label='Space (bytes)')
plt.title('Time vs Space Complexity of Merge Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# ----------------------------
# Written Analysis
# ----------------------------
print("\n--- Merge Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")
print("\nTime Complexity:")
print("  Best Case: O(n log n)")
print("  Average Case: O(n log n)")
print("  Worst Case: O(n log n)")
print("\nSpace Complexity:")
print("  O(n) due to temporary arrays used in merging")
```

```python
#Quick Sort
import time
import matplotlib.pyplot as plt
import sys
import random

# Quick Sort Implementation
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


# Changed input sizes
input_sizes = [10, 100, 500, 1000, 5000, 10000]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    # Measure execution time
    start = time.time()
    quick_sort(arr)
    end = time.time()
    execution_times.append(end - start)

    # Measure space usage (approx)
    space_usage.append(sys.getsizeof(arr) + (size * sys.getsizeof(arr[0])))


# Plotting
plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='red', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='purple', label='Space (bytes)')
plt.title('Time vs Space Complexity of Quick Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()


# Written Analysis
print("\n--- Quick Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")
print("\nTime Complexity:")
print("  Best Case: O(n log n) when pivot divides array evenly")
print("  Average Case: O(n log n)")
print("  Worst Case: O(n^2) when pivot is smallest/largest element repeatedly")
print("\nSpace Complexity:")
print("  O(log n) for recursion stack in best/average cases")
print("  O(n) in worst case recursion depth")
```

```python
#Insertion Sort
import time
import matplotlib.pyplot as plt
import sys
import random


# Insertion Sort Implementation
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr


# Changed values
input_sizes = [10, 20, 50, 100, 200, 400]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    # Measure execution time
    start = time.time()
    insertion_sort(arr.copy())
    end = time.time()
    execution_times.append(end - start)

    # Measure space usage (array + elements)
    space_usage.append(sys.getsizeof(arr) + (size * sys.getsizeof(arr[0])))


# Plotting
plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='green', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='orange', label='Space (bytes)')
plt.title('Time vs Space Complexity of Insertion Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()


# Written Analysis
print("\n--- Insertion Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")
print("\nTime Complexity:")
print("  Best Case: O(n) when array is already sorted")
print("  Average Case: O(n^2)")
print("  Worst Case: O(n^2) when array is reverse sorted")
print("\nSpace Complexity:")
print("  O(1)  In-place sorting (no extra space needed)")
```

```python
#Bubble Sort
import time
import random
import sys
import matplotlib.pyplot as plt

# Bubble Sort Implementation
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Changed sizes
sizes = [100, 200, 400, 600, 800, 1000, 1500]
time_taken = []
space_used = []

for size in sizes:
    arr = [random.randint(1, 10000) for _ in range(size)]

    space_used.append(sys.getsizeof(arr))

    start = time.time()
    bubble_sort(arr.copy())
    end = time.time()

    time_taken.append(end - start)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(sizes, time_taken, marker='o', color='blue', label='Execution Time')
ax1.set_xlabel('Array Size (n)')
ax1.set_ylabel('Execution Time (seconds)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(sizes, space_used, marker='x', color='orange', linestyle='--', label='Space Used')
ax2.set_ylabel('Space (bytes)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Time and Space Complexity of Bubble Sort')
plt.grid(True)
plt.show()

print("\n--- Bubble Sort Analysis ---")
print("Input: An unsorted list of numbers")
print("Output: A sorted list of numbers (ascending order)")

print("\nTime Complexity:")
print("  - Best Case: O(n) [when the array is already sorted]")
print("  - Average Case: O(n^2)")
print("  - Worst Case: O(n^2) [when the array is sorted in reverse order]")

print("\nSpace Complexity: O(1) (in-place sorting, only a few extra variables needed)")
```

```python
#Selection Sort
import time
import matplotlib.pyplot as plt
import sys
import random

# Selection Sort Implementation
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Changed input sizes
input_sizes = [10, 100, 300, 500, 700, 1000]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    start = time.time()
    selection_sort(arr.copy())
    end = time.time()
    execution_times.append(end - start)

    space_usage.append(sys.getsizeof(arr) + (size * sys.getsizeof(arr[0])))

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='red', label='Space (bytes)')
plt.title('Time vs Space Complexity of Selection Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Selection Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")

print("\nTime Complexity:")
print("  Best Case: O(n^2) (no early termination possible)")
print("  Average Case: O(n^2)")
print("  Worst Case: O(n^2)")

print("\nSpace Complexity:")
print("  O(1)  In-place sorting (no extra space needed)")
```

```python
#Binary Search
import time
import matplotlib.pyplot as plt
import sys
import random

# Binary Search Implementation
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Changed input sizes
input_sizes = [10, 1000, 5000, 10000, 50000, 100000, 200000]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = sorted([random.randint(1, 1000000) for _ in range(size)])
    target = random.choice(arr)

    start = time.time()
    binary_search(arr, target)
    end = time.time()
    execution_times.append(end - start)

    space_usage.append(sys.getsizeof(arr) + (size * sys.getsizeof(arr[0])))

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='red', label='Space (bytes)')
plt.title('Time vs Space Complexity of Binary Search')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Binary Search Analysis ---")
print("Input/Output:")
print("  Input: Sorted array of n integers and a target element")
print("  Output: Index of target element (or -1 if not found)")

print("\nTime Complexity:")
print("  Best Case: O(1)  Target found at middle on first try")
print("  Average Case: O(log n)")
print("  Worst Case: O(log n)  Target not found or at extreme end")

print("\nSpace Complexity:")
print("  Iterative: O(1)  Constant extra space")
print("  Recursive: O(log n)  Due to recursion stack")
```
